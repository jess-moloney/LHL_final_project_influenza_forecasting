### Cleaning the Data

```python
# get health indicators
df_health_indicators = pd.read_csv(r'..\data\health_indicators\health_indicators_2015_2021.csv', skiprows=12)

# Drop nulls because they are only showing on the rows without data
df_health_indicators.dropna(inplace=True)

# Null values in the data are represented with '..'
df_health_indicators = df_health_indicators.replace('..', np.nan)
df_health_indicators = df_health_indicators.T
more_than_half_missing = df_health_indicators.isnull().sum() >= 4
more_than_half_missing = pd.DataFrame(more_than_half_missing)
idx_to_drop = list(more_than_half_missing[more_than_half_missing[0]==True].index.values)

# drop rows
df_health_indicators = df_health_indicators.T
df_health_indicators = df_health_indicators.drop(index=idx_to_drop)
df_health_indicators = df_health_indicators.reset_index(drop=True)

# convert values to numeric values
df_health_indicators = df_health_indicators.replace(',','', regex=True)
df_health_indicators = df_health_indicators.set_index('Indicators')
df_health_indicators = df_health_indicators.apply(pd.to_numeric)

# fill nulls with the mean value from the row
df_health_indicators = df_health_indicators.apply(lambda row: row.fillna(row.mean()), axis=1)
df_health_indicators.reset_index()
df_health_indicators = df_health_indicators.T
df_health_indicators = df_health_indicators.reset_index()
df_health_indicators.rename(columns={'index':'year'},inplace=True)
df_health_indicators['year'] = df_health_indicators['year'].astype(int)

# check the length of the dataframe, check for duplicates, check for nulls
print(f"Number of years: {len(df_health_indicators)}")
print(f'Duplicates: {df_health_indicators.duplicated().any()}')
print(f'Nulls:\n{df_health_indicators.isnull().sum()}')
df_health_indicators.head()
```

```python
# add a row for 2022 and shift all the values forward one year because we will use previous year's value to predict current years flu cases
df_health_indicators.loc[len(df_health_indicators.index)] = 2022
df_health_indicators.iloc[:,1:] = df_health_indicators.iloc[:,1:].shift(1)

pickle.dump(df_health_indicators, open(r"..\data\health_indicators\df_health_indicators.pkl", "wb" ))
```

### Cleaning the Data

```python
# drop NaNs because they are only showing on the rows without data
df_air_travel.dropna(inplace=True)
df_air_travel = df_air_travel.reset_index(drop=True)

# null values in the data are represented with '..'
df_air_travel = df_air_travel.replace('..', np.nan)

# remove all rows except the domestic and international air passengers 
df_air_travel = df_air_travel.iloc[[0,1,6],:].reset_index(drop=True)

# convert values to numeric values
df_air_travel = df_air_travel.replace(',','', regex=True)
df_air_travel = df_air_travel.set_index('Activity indicators')
df_air_travel = df_air_travel.apply(pd.to_numeric)

# get DateTime Index
df_air_travel = df_air_travel.T
df_air_travel.index = pd.to_datetime(df_air_travel.index)

# rename columns
df_air_travel.columns = ['Air Passengers between Canada & US (x1000)',
                                'Air Passengers between Canada & Overseas (x1000)',
                                'All Air Passengers (x1000)']

# check the length of the dataframe, check for duplicates, check for nulls
print(f"Number of weeks: {len(df_air_travel)}")
print(f'Duplicates: {df_air_travel.duplicated().any()}')
print(f'Nulls:\n{df_air_travel.isnull().sum()}')
df_air_travel.head()
```

```python
# shift all the values forward one month because we will use previous month's value to predict current months flu cases
df_air_travel.iloc[:,:] = df_air_travel.iloc[:,:].shift(1)

# extract month and year from index for merging with other dataframes
years = [x.year for x in df_air_travel.index]
months = [x.month for x in df_air_travel.index]

X = np.array([years, months]).T
dates = pd.DataFrame(X, columns = ['year', 'month'])

df_air_travel = df_air_travel.reset_index(drop=True)
df_air_travel = pd.concat([dates, df_air_travel], axis=1)

pickle.dump(df_air_travel, open(r"..\data\mobility\df_air_travel.pkl", "wb" ))
```

# Google Trends - Flu Symptom Search Terms (Monthly, 2015-Present)

### Loading and Examining the Data

```python
# get a list of table names
table_list_flu = os.listdir(r"..\data\google_trend\flu")

# create a dictionary with table names as keys and dataframes as values
dict_flu = {}
datapath = r'..\data\google_trend\flu'

for i in range(len(table_list_flu)):
    df = pd.read_csv(datapath+'\\'+table_list_flu[i], header=1)
    dict_flu[table_list_flu[i][13:-4]] = df

dict_flu['cough']
```

### Cleaning the Data

```python
# create a dataframe with one column for each symptom
df_flu_symptoms = pd.DataFrame({'Month':dict_flu['body_aches']['Month']}) # get the month values from one of the dataframes

for key in dict_flu:
    df_flu_symptoms[key] = dict_flu[key].iloc[:,1] # retrieve values from each dataframe

# check the length of the dataframe, check for duplicates, check for nulls
print(f"Number of months: {len(df_flu_symptoms)}")
print(f'Duplicates: {df_flu_symptoms.duplicated().any()}')
print(f'Nulls:\n{df_flu_symptoms.isnull().sum()}')
df_flu_symptoms.head()
```

```python
# add extra month at the end of dataframe to allow for future forecast and convert values to DateTime
df_flu_symptoms.loc[len(df_flu_symptoms.index)] = '2022-12'
df_flu_symptoms['Month'] = pd.to_datetime(df_flu_symptoms['Month'])
df_flu_symptoms.set_index('Month', inplace=True)

# shift all the values forward one month because we will use previous month's value to predict current months flu cases
df_flu_symptoms.iloc[:,1:] = df_flu_symptoms.iloc[:,1:].shift(1)

# get month and year to be able to merge with df_flu
months = [x.month for x in df_flu_symptoms.index]
years = [x.year for x in df_flu_symptoms.index]
X = np.array([months, years]).T
dates = pd.DataFrame(X, columns = ['month', 'year'])
df_flu_symptoms = df_flu_symptoms.reset_index(drop=True)
df_flu_symptoms = pd.concat([dates, df_flu_symptoms], axis=1)

pickle.dump(df_flu_symptoms, open(r"..\data\google_trend\df_flu_symptoms.pkl", "wb" ))
```

```python
df_flu = pd.merge(df_flu, df_health_indicators, how='left', on='year')
df_flu = pd.merge(df_flu, df_air_travel, how='left', on=['year','month'])
df_flu = pd.merge(df_flu, df_urban_transit, how='left', on=['year','month'])
df_flu = pd.merge(df_flu, df_flu_symptoms, how='left', on=['month','year'])
df_flu = pd.merge(df_flu, df_COVID_terms, how='left', on=['month','year'])
```
